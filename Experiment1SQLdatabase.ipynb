```python
import pandas as pd
import sqlite3

# Load the CSV data into a Pandas DataFrame
file_path = '/Users/stephaniebellew/Desktop/Experiment1_Location1.csv'
df = pd.read_csv(file_path)

# Fill missing values with 'missing' as you already did
df = df.fillna('missing')

# Connect to SQLite database (it will create the file if it doesn't exist)
db_path = '/Users/stephaniebellew/Desktop/Ventilation_Experiments_Metadata.db'
conn = sqlite3.connect(db_path)

# Convert the DataFrame into an SQL table named 'Ventilation_Experiments'
df.to_sql('Ventilation_Experiments_Metadata', conn, if_exists='replace', index=False)

# Verify by fetching some data from the database
query = "SELECT * FROM Ventilation_Experiments LIMIT 200;"
result = pd.read_sql(query, conn)

# Display the result (first 5 rows of the table)
print(result)

# Close the connection to the database
conn.close()
```

```python
import pandas as pd
import sqlite3

# Path to the second CSV file
file_path_2 = '/Users/stephaniebellew/Desktop/Experiment1_Location2.csv'

# Load the second CSV into a DataFrame
df_2 = pd.read_csv(file_path_2)

# Fill missing values with 'missing' (same as before)
df_2 = df_2.fillna('missing')

# Connect to the existing SQLite database
db_path = '/Users/stephaniebellew/Desktop/Ventilation_Experiments.db'
conn = sqlite3.connect(db_path)

# Append the data from the second CSV to the existing SQL table
df_2.to_sql('Ventilation_Experiments', conn, if_exists='append', index=False)

# Verify by fetching some data from the table
query = "SELECT * FROM Ventilation_Experiments LIMIT 40000;"



# Execute the query and load the result into a DataFrame
result = pd.read_sql(query, conn)

# Display the result (first 5 rows of the table)
print(result)

#Query to select rows where 'Overall Id' contains '1_1'
query = "SELECT * FROM Ventilation_Experiments WHERE `Overall Id` LIKE '%1_1%';"

# Execute the query and load the result into a DataFrame
result = pd.read_sql(query, conn)

# Display the result (first 5 rows of the table)
print(result.head())


# Close the database connection
conn.close()
```
