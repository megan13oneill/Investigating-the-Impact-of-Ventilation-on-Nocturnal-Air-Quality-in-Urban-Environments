```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
from matplotlib.patches import FancyBboxPatch

# Step 1: Load the CSV data
df = pd.read_excel('/Users/stephaniebellew/Desktop/IAQWC.xlsx')

# Step 2: Check if the column 'IAQ Difference' exists
if 'IAQ Difference' not in df.columns:
    raise ValueError("Column 'IAQ Difference' not found in the CSV file.")

# Step 3: Calculate key statistics for the IAQ Difference
iqd = df['IAQ Difference']

# Median
median_val = np.median(iqd)

# Quartiles (Q1, Q3)
Q1 = np.percentile(iqd, 25)
Q3 = np.percentile(iqd, 75)

# Interquartile range (IQR)
IQR = Q3 - Q1

# Outliers (fliers): values outside 1.5 * IQR from the quartiles
outliers = iqd[(iqd < (Q1 - 1.5 * IQR)) | (iqd > (Q3 + 1.5 * IQR))]

# Mean (average) value
mean_val = np.mean(iqd)

# Step 4: Create the box plot with bright dark orange fill and dark blue lines
sns.set(style="whitegrid")  # Set plot style
plt.figure(figsize=(8, 6))   # Set plot size

box = sns.boxplot(x=iqd,
                  color='#FF6600',           # Box fill color set to bright dark orange
                  fliersize=5,               # Size of outliers
                  linewidth=2,               # Line width for box plot elements
                  whis=1.5,                  # Whisker range (adjust if necessary)
                  palette=["#FF6600"])       # Ensure the color is the bright dark orange (#FF6600)

# Explicitly change the colors of all components (whiskers, caps, medians, etc.)
for line in box.lines:
    if 'whisker' in line.get_label():
        line.set_color('darkblue')  # Whiskers to dark blue
    elif 'median' in line.get_label():
        line.set_color('darkblue')  # Median line to dark blue
    elif 'cap' in line.get_label():
        line.set_color('darkblue')  # Caps to dark blue
    line.set_linewidth(2)  # Increase line width for better visibility

# Set the box borders to dark blue
for artist in box.artists:
    artist.set_edgecolor('darkblue')  # Box edges to dark blue
    artist.set_linewidth(2)  # Set the width of the box borders

# Set the outlier points (fliers) to dark blue edges and bright dark orange fill
for flier in box.collections:
    flier.set_edgecolor('darkblue')  # Set outlier border to dark blue
    flier.set_facecolor('#FF6600')   # Set outlier fill color to bright dark orange

# Step 5: Annotate key statistics with a box around them
plt.title('Comparing the Impact of Ventilation on IAQ Across Two Locations', fontsize=16, color='black')  # Title color in black
plt.xlabel('Difference in IAQ Between Arduino 1 and Arduino 2', fontsize=14, color='black')  # Axis labels in default color (black)
plt.ylabel('')  # Remove Y-label as it isn't necessary here

# Add a box around the key statistics
key_stats_box = FancyBboxPatch((0.02, 0.65), 0.15, 0.3, boxstyle="round,pad=0.1", linewidth=2, facecolor="lightgray", edgecolor="black", alpha=0.8, transform=plt.gca().transAxes)
plt.gca().add_patch(key_stats_box)

# Display the key statistics in the box
plt.text(0.05, 0.90, f'Median: {median_val:.2f}', ha='left', va='top', fontsize=12, color='darkblue', transform=plt.gca().transAxes)
plt.text(0.05, 0.85, f'Mean: {mean_val:.2f}', ha='left', va='top', fontsize=12, color='darkblue', transform=plt.gca().transAxes)
plt.text(0.05, 0.80, f'Q1: {Q1:.2f}', ha='left', va='top', fontsize=12, color='darkblue', transform=plt.gca().transAxes)
plt.text(0.05, 0.75, f'Q3: {Q3:.2f}', ha='left', va='top', fontsize=12, color='darkblue', transform=plt.gca().transAxes)
plt.text(0.05, 0.70, f'IQR: {IQR:.2f}', ha='left', va='top', fontsize=12, color='darkblue', transform=plt.gca().transAxes)

if len(outliers) > 0:
    plt.text(0.05, 0.65, f'Outliers: {len(outliers)}', ha='left', va='top', fontsize=12, color='darkblue', transform=plt.gca().transAxes)

# Step 6: Save the plot to the desktop
desktop_path = os.path.expanduser('~/Desktop')  # Get the path to the desktop
file_path = os.path.join(desktop_path, 'IAQ_Difference_BoxPlot_Annotatedexp1.png')  # Define the full file path

# Save the plot
plt.savefig(file_path, bbox_inches='tight')  # Save the plot with tight layout

# Step 7: Display the plot
plt.show()

# Print message to confirm the file is saved
print(f"Plot saved to {file_path}")
```
